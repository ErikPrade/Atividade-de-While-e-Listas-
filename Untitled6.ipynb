```python
vendas = []

for i in range(5):
    valor = float(input(f"Digite o valor da venda {i+1}: "))
    vendas.append(valor)

total_vendido = sum(vendas)
media = total_vendido / 5

acima_da_media = 0
for venda in vendas:
    if venda > media:
        acima_da_media = acima_da_media + 1

print("Resultados")
print("Total vendido: R$", total_vendido)
print("Média das vendas: R$", media)
print("Quantidade de vendas acima da média:", acima_da_media)
```

```python
funcionarios = []

while True:
    nome = input("Digite o nome do funcionário (ou 'sair' para finalizar): ")
    if nome == 'sair':
        break
    funcionarios.append(nome)

print("Funcionários Cadastrados")
for funcionario in funcionarios:
    print(funcionario)

print("Quantidade total de funcionários:", len(funcionarios))
```

```python
produtos = ["Teclado", "Mouse", "Monitor", "Cabo de rede", "Webcam"]
estoques = [12, 3, 5, 2, 8]

precisa_reposicao = 0

print("Produtos com estoque menor que 5:")
for i in range(len(produtos)):
    if estoques[i] < 5:
        print("Produto:", produtos[i], "| Estoque:", estoques[i])
        precisa_reposicao = precisa_reposicao + 1

print("Total de produtos que precisam de reposição:", precisa_reposicao)
```

```python
compras = []
total_gasto = 0

while total_gasto <= 5000:
    valor = float(input("Digite o valor da compra: R$ "))
    compras.append(valor)
    total_gasto = total_gasto + valor

print("Controle de Gastos")
print("Total gasto: R$", total_gasto)
print("Quantidade de compras realizadas:", len(compras))
```

```python
tempos_atendimento = [15, 45, 20, 35, 10, 60]

total_tempo = sum(tempos_atendimento)
quantidade = len(tempos_atendimento)
media = total_tempo / quantidade

print("Tempo total de atendimento:", total_tempo, "minutos")
print("Média de tempo por chamado:", media, "minutos")

print("Chamados que demoraram mais de 30 minutos:")
for tempo in tempos_atendimento:
    if tempo > 30:
        print(tempo, "minutos")
```
